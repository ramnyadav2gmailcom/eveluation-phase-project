# predict the capacity_mw 

```python
x_a=df.drop('capacity_mw',axis=1)
```

```python
y_b=df.capacity_mw
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_a=pd.DataFrame(sc.fit_transform(x_a),columns=x_a.columns)
x_a
```

```python
train_xa,test_xa,train_yb,test_yb=train_test_split(x_a,y_b,test_size=0.33,random_state=43)
```

```python
train_xa.shape
```

```python
train_yb.shape
```

```python
test_xa.shape
```

```python
test_yb.shape
```

```python
lr=LinearRegression()
lr.fit(train_xa,train_yb)
pred_lr=lr.predict(test_xa)
pred_train=lr.predict(train_xa)
print('R2 score:',r2_score(test_yb,pred_lr))
print("r2 score on training data :",r2_score(train_yb,pred_train))
print("mean absolute error:",mean_absolute_error(test_yb,pred_lr))
print("mean squared error:",mean_squared_error(test_yb,pred_lr))
print("root mean squared error:",np.sqrt(mean_squared_error(test_yb,pred_lr)))
```

```python
from sklearn.ensemble import ExtraTreesRegressor
etr=ExtraTreesRegressor()
etr.fit(train_xa,train_yb)
etr_pred=etr.predict(test_xa)
etr_train=etr.predict(train_xa)
print('R2 score:',r2_score(test_yb,etr_pred))
print('R2  score on training data:',r2_score(train_yb,etr_train))
print('mean absolute error:',mean_absolute_error(test_yb,etr_pred))
print('mean squared errro:',mean_squared_error(test_yb,etr_pred))
print('root mean squared error:',np.sqrt(mean_squared_error(test_yb,etr_pred)))
    
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr=GradientBoostingRegressor()
gbr.fit(train_xa,train_yb)
gbr_pred=gbr.predict(test_xa)
gbr_train=gbr.predict(train_xa)
print('R2 score:',r2_score(test_yb,gbr_pred))
print('R2  score on training data:',r2_score(train_yb,gbr_train))
print('mean absolute error:',mean_absolute_error(test_yb,gbr_pred))
print('mean squared errro:',mean_squared_error(test_yb,gbr_pred))
print('root mean squared error:',np.sqrt(mean_squared_error(test_yb,gbr_pred)))
    
```

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
rd=Ridge()
rd.fit(train_xa,train_yb)
pred_rd=rd.predict(test_xa)
pred_train=rd.predict(train_xa)
print('R2 score',r2_score(test_yb,pred_rd))
print('R2 score on training data',r2_score(train_yb,pred_train))
print('mean absolute errot',mean_absolute_error(test_yb,pred_rd))
print('mean squared error',mean_squared_error(test_yb,pred_rd))
print('root mean squared errror',np.sqrt(mean_squared_error(test_yb,pred_rd)))
```

```python
ls=Lasso()
ls.fit(train_xa,train_yb)
pred_ls=rd.predict(test_xa)
pred_train=ls.predict(train_xa)
print('R2 score',r2_score(test_yb,pred_ls))
print('R2 score on training data',r2_score(train_yb,pred_train))
print('mean absolute errot',mean_absolute_error(test_yb,pred_ls))
print('mean squared error',mean_squared_error(test_yb,pred_ls))
print('root mean squared errror',np.sqrt(mean_squared_error(test_yb,pred_ls)))
```
